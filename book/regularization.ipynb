# Model-based and regularization approach

**How to cure ill-posedness?**

1. By searching approximate solutions satisfying additional constraints 
deriving from the physics of the problem.

The information added is also called *a priori or prior image* information.

In mathematical models this a priori information is called *regularization*.

It is added by means of one or more regularization terms, each multplied by a positive scalar called regularization parameter.

The simplest regularization model with a single regularization function is expressed as:
$$ min \|Ax-(y+e)\|_2^2 + \lambda R(x)$$

Where:

- $R(x)$ is the regularization function that incrporates a priori infromation on the solution

- $\lambda$ is the regularization parameter representing the trade off between the fit-to-data  and the regularization terms.

In the following, we will first consider the previuous model with different possible regularization functions $R(x)$.


**Model-based approach**

We speak in this case of direct regularization.

*Direct regularization methods* compute the solution of an ill posed inverse problem (imaging)  by imposing some constraints on the solution. 

The so called model-based approach mathematically models the problem to be solved as a minimization problem:
$$ min \|Ax-(y+e)\|_2^2 + \lambda R(x)$$

 with two acting functions: 

- The term   $\|Ax-(y+e)\|_2^2=\|Ax-y^{\delta}\|_2^2$  representing the fit with the data, i.e. imposing coherence of the solution with the measured data              

- The regularization term R(x) that incorporates a priori information on the solution.

The model can also be formulated as a constrained minimization in one of the following forms:

$min_x  R(x)$ such that $\|Ax-y^{\delta}\|_2^2=\epsilon$

$min \|Ax-y^{\delta}\|_2^2$ such that $R(x)= \sigma$

In the first formulation we suppose to have an estimate of the residual norm \|Ax-y^{\delta}\|_2^2$, whereas in the second formulation we suppose to have an estimate of the value of $R(x)$.
